```python
from histpy import Histogram
import numpy as np
```

```python
h2D = Histogram([np.arange(0,1411,1), np.arange(0,1604,1)], labels = ['x','y'])
```

```python
for x_idx in tqdm(np.arange(0,1411,1)):
    for y_idx in np.arange(0,1604,1):
        h2D[x_idx, y_idx] = data[x_idx, y_idx]
```

```python
h2D.draw()
```
